```python
from transformers import AutoTokenizer
import transformers
import torch

model = "MrShaheer/Recipe-llama-2-7b-custom-2k"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
system_message  = "Welcome to the Ingredient-Bound Recipe Generator! To generate a recipe, enter the exact ingredients you have, separated by commas (e.g., chicken, carrots, onions, rice). Remember, no substitutions or additions are permitted. Once you've listed all ingredients, type 'GENERATE!' and I'll prepare a delicious recipe just for you. For any other inquiries, please consult the provided documentation or contact support. Now, let's get started! What ingredients can you offer?"
```

```python
import torch

# Release GPU memory
torch.cuda.empty_cache()
```

```python
prompt = f"[INST] <>\n{system_message}\n<>\n\n onions,potato,rice,tomato,garlic,oil [/INST]"
sequences = pipeline(
    f'[INST] {prompt} [/INST]',
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=800,
)
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```

```python
prompt = f"[INST] <>\n{system_message}\n<>\n\n onions,tomatoes,capsicum,yogurt [/INST]"
sequences = pipeline(
    f'[INST] {prompt} [/INST]',
    do_sample=True,
    top_k=10,
    num_return_sequences=2,
    eos_token_id=tokenizer.eos_token_id,
    max_length=400,
)
c = 0
for seq in sequences:
    if c == 0:
      st = seq['generated_text']

      print(f"{st[len(prompt) + 16: ]}")
      c += 1
    else:
      print(f"Result: {seq['generated_text']}")
```
